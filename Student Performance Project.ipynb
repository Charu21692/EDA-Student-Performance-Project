### **5.1 Handling of Missing Data**

- In this section, we will identify missing data and check the proportion of it and take appropriate measures.

```python
round((100*(SP.shape[0] - SP.count())/SP.shape[0]).sort_values(ascending = False),2)
# or
# round((100 * SP.isnull().sum()/SP.shape[0]),2)
```

**Observations:**

- We can observe that following features are found to have missing values along with the proportions:

|Feature|Object Type|Missing Proportion|Solution|
|:--:|:--:|:--:|:--|
|TransportMeans|Object|10.23%|Replace with unknowm.|
|ParentEduc|Object|6.02%|Replace with mode.|
|EthnicGroup|Object|6.01%|Replace with mode.|
|TestPrep|Object|5.97%|Replace with mode.|
|NrSiblings|float64|5.13%|Replace with unknowm.|
|ParentMaritalStatus|Object|3.88%|Replace with unknowm.|
|WklyStudyHours|Object|3.12%|Replace with mode.|
|IsFirstChild|Object|2.95%|Replace with mode.|
|PracticeSport|Object|2.06%|Replace with mode.|

**Correction for Missing Values**

```python
SP['EthnicGroup'] = SP['EthnicGroup'].replace(np.nan, SP['EthnicGroup'].mode()[0])

SP['ParentEduc'] = SP['ParentEduc'].replace(np.nan, SP['ParentEduc'].mode()[0])

SP['TestPrep'] = SP['TestPrep'].replace(np.nan, SP['TestPrep'].mode()[0])

SP['ParentMaritalStatus'] = SP['ParentMaritalStatus'].replace(np.nan, "Unknown")

SP['PracticeSport'] = SP['PracticeSport'].replace(np.nan, SP['PracticeSport'].mode()[0])

SP['IsFirstChild'] = SP['IsFirstChild'].replace(np.nan, SP['IsFirstChild'].mode()[0])

SP['TransportMeans'] = SP['TransportMeans'].replace(np.nan, "Unknown")
```

```python
SP['WklyStudyHours'].value_counts()    # Handling of inconsistant data
```

```python
SP['WklyStudyHours'] = SP['WklyStudyHours'].replace('5-Oct', "5-10")
```

```python
SP['WklyStudyHours'].value_counts()
```

```python
SP['WklyStudyHours'] = SP['WklyStudyHours'].replace(np.nan, SP['WklyStudyHours'].mode()[0])
```

```python
round(SP['NrSiblings'].mean())
```

```python
SP['NrSiblings'] = SP['NrSiblings'].replace(np.nan, round(SP['NrSiblings'].mean()))
```
